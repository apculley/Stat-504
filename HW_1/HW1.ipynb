Exercise 1: Write a program that reads a file, breaks each line into words, strips whitespace and punctuation from the words, and converts them to lowercase. 

```python
def txt_prep(filepath):
    out=open(filepath).read()
    out=out.lower()
    b=out.replace("-"," ")
 
    from string import punctuation
    out=[]
    for word in b.split():
        word=word.strip()
        word=word.strip(punctuation)
        out.append(word)
    return out
txt_prep('/home/thyme/Downloads/monte cristo.txt')
```

This function processes the txt file into a list of words we're trying to avoid:

```python
def process_forbidden(filepath):
    a=open(filepath).read()
    a=a.lower()
    b=a.replace("\n"," ")
    forbidden=[]
    from string import punctuation
    for word in b.split():
        word=word.strip()
        word=word.strip(punctuation)
        forbidden.append(word)
    i=0
    for word in forbidden:
        forbidden[i]=word.replace("\xc2\xa0","")
        i=i+1
    forbidden=list(set(forbidden))
    return forbidden
```

This function removes the boring words from our dictionary of words used over 100 times.

```python
def remove_words(t,a,b,forbidden):
    wordlist=[]
    for item in t:
        for word in forbidden:
            if word in item:
                wordlist.append(item)
    print  "Word count minus boring words:\n"+str(a-sum(x[0] for x in wordlist))   
    final=[]
    for item in t:
        if item not in wordlist:
            final.append(item)
    print "Unique word count minus boring words:\n"+str(b-len(wordlist))
    return final
```
